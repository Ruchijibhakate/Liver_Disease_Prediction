### Domain Analysis

#### 1. **Age**
- **Impact**: Older age increases risk due to prolonged exposure to liver disease risk factors.
- **Relevance**: Helps identify age-specific trends in liver disease prevalence and severity.

#### 2. **Gender**
- **Impact**: Men generally have higher risk due to behaviors like alcohol consumption. Hormonal differences also affect liver disease progression.
- **Relevance**: Essential for understanding gender-specific risk and tailoring interventions.

#### 3. **Total Bilirubin**
- Total bilirubin is the sum of both unconjugated (indirect) bilirubin and conjugated (direct) bilirubin in the blood. It reflects the overall level of bilirubin, which is a byproduct of the normal breakdown of red blood cells.
- **Impact**: High levels indicate liver dysfunction or bile duct issues.
- **Relevance**: Critical for diagnosing jaundice and monitoring liver health.

#### 4. **Direct Bilirubin**
- Direct bilirubin is the bilirubin that has been processed by the liver and is water-soluble. It is measured directly in the blood and provides insight into the liver's ability to process and excrete bilirubin.
- **Impact**: Elevated levels suggest liver or bile duct obstruction.
- **Relevance**: Useful for differentiating between types of jaundice.

#### 5. **Alkaline Phosphatase (ALP)**
- Alkaline phosphatase (ALP) is an enzyme found in many tissues throughout the body, with high concentrations in the liver, bile ducts, bones, and placenta. It plays a crucial role in breaking down proteins and is important for various bodily functions, including bone formation and bile production.
- **Impact**: High levels can indicate liver disease, bile duct obstruction, or bone disease.
- **Relevance**: Important for diagnosing liver and bile duct disorders.

#### 6. **Alanine Aminotransferase (ALT)**
- Alanine Aminotransferase (ALT) is an enzyme found primarily in the liver, but also in smaller amounts in the kidneys and muscles. It plays a key role in the metabolism of amino acids and is often measured to assess liver health.    
- **Impact**: High levels signal liver damage or inflammation.
- **Relevance**: Widely used to assess liver function and detect liver disease.

#### 7. **Aspartate Aminotransferase (AST)**
- Aspartate Aminotransferase (AST), also known as Serum Glutamic-Oxaloacetic Transaminase (SGOT), is an enzyme found primarily in the liver, but also in the heart, kidneys, brain, and muscles. It plays a role in amino acid metabolism and is often measured to assess liver health.
- **Impact**: Elevated levels indicate liver injury; often used with ALT for a comprehensive assessment.
- **Relevance**: Helps evaluate liver health and detect damage.

#### 8. **Total Proteins**
- Total Protein levels in the blood are an important indicator of overall health, including liver function.
- **Impact**: Low levels can indicate liver disease, kidney disease, or nutritional issues.
- **Relevance**: Important for assessing overall liver function and nutritional status.

#### 9. **Albumin**
- Albumin is the most abundant protein in the blood, produced by the liver. It helps maintain oncotic pressure (preventing fluid leakage from blood vessels), transports hormones, vitamins, and drugs, and provides nourishment to tissues.    
- **Impact**: Low levels suggest chronic liver disease or severe liver damage.
- **Relevance**: Key marker for liver function and nutritional status.

#### 10. **Albumin and Globulin Ratio**
- The Albumin-to-Globulin (A/G) Ratio is a blood test that measures the relative amounts of albumin and globulin in blood. This ratio helps assess liver function, nutritional status, and immune system activity.    
- **Impact**: Altered ratios can indicate liver disease or immune disorders.
- **Relevance**: Useful for diagnosing liver disease and monitoring immune function.

### Conclusion
Each of these features plays a critical role in diagnosing, monitoring, and understanding liver disease. They provide a comprehensive view of a patient's liver health, helping healthcare providers make informed decisions.

### 5. Outlier Handling

#### Creating a function that calculates the IQR and then filters out the outliers.

```python
def remove_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    median_value = data[column].median() 
    data[column] = data[column].apply(lambda x: median_value if x < lower_bound or x > upper_bound else x) 
    return sns.boxplot(x=column,data=data)
```

## **Model Performance Evaluation**

In this project, multiple metrics were employed to evaluate the performance of predictive models for liver disease diagnosis. These metrics include accuracy, F1 score, recall, precision, classification report, cross-validation, and ROC AUC curve. Below, each metric is explained in detail, including its importance and interpretation.

#### 1. **Accuracy**
**Definition**: Accuracy is the ratio of correctly predicted instances to the total instances. It provides an overall measure of how well the model performs across all classes.

**Formula**: Accuracy = (True Positives + True Negatives) / (Total Predictions)

**Importance**: Accuracy gives a quick overview of the model's performance but may not be sufficient alone, especially in cases of imbalanced datasets.

**Interpretation**: Higher accuracy indicates better performance, but it should be interpreted alongside other metrics.

#### 2. **F1 Score**
**Definition**: The F1 score is the harmonic mean of precision and recall, providing a single metric that balances both concerns.

**Formula**: F1 Score = 2 * (Precision * Recall) / (Precision + Recall)

**Importance**: The F1 score is particularly useful when dealing with imbalanced datasets, as it considers both false positives and false negatives.

**Interpretation**: A higher F1 score indicates a good balance between precision and recall, which is crucial for medical diagnosis where both false positives and false negatives can have serious consequences.

#### 3. **Recall (Sensitivity)**
**Definition**: Recall is the ratio of correctly predicted positive observations to all actual positives. It measures the model's ability to identify true positives.

**Formula**: Recall = True Positives / (True Positives + False Negatives).

**Importance**: High recall is essential in medical diagnostics to minimize false negatives, ensuring that most patients with the disease are correctly identified.

**Interpretation**: Higher recall means fewer false negatives, which is critical in detecting liver disease early.

#### 4. **Precision**
**Definition**: Precision is the ratio of correctly predicted positive observations to the total predicted positives. It measures the accuracy of positive predictions.

**Formula**: Precision = True Positives / (True Positives + False Positives)

**Importance**: Precision is important to minimize false positives, ensuring that non-diseased patients are not incorrectly diagnosed.

**Interpretation**: Higher precision means fewer false positives, leading to fewer unnecessary treatments or interventions.

#### 5. **Classification Report**
**Definition**: A classification report provides a detailed performance analysis of the model by reporting precision, recall, and F1 score for each class.

**Importance**: It gives a comprehensive overview of the model's performance across different classes, which is crucial for understanding how well the model performs for each category.

**Interpretation**: The classification report helps in identifying which classes are well-predicted and which need improvement.

#### 6. **Cross-Validation**
**Definition**: Cross-validation involves dividing the dataset into multiple folds and training the model on each fold while validating on the remaining data. This process is repeated multiple times to ensure robustness.

**Importance**: Cross-validation provides a more reliable estimate of the model's performance by minimizing overfitting and ensuring that the model generalizes well to unseen data.

**Interpretation**: Consistently high performance across different folds indicates a robust and reliable model.

#### 7. **ROC AUC Curve**
**Definition**: The ROC (Receiver Operating Characteristic) curve plots the true positive rate (recall) against the false positive rate at various threshold settings. The AUC (Area Under the Curve) measures the entire two-dimensional area underneath the ROC curve.

**Importance**: The ROC AUC is a comprehensive measure of model performance, indicating how well the model distinguishes between positive and negative classes.

**Interpretation**: An AUC of 0.5 indicates random performance, whereas an AUC closer to 1 indicates excellent performance.

### Conclusion
The combination of these metrics provides a holistic view of model performance. Accuracy offers a general overview, while precision and recall give insights into the balance between false positives and false negatives. The F1 score combines precision and recall into a single metric, and the classification report breaks down these metrics by class. Cross-validation ensures the model's robustness, and the ROC AUC curve provides a comprehensive assessment of the model's ability to distinguish between classes. Together, these metrics enable a thorough evaluation of the predictive model's effectiveness in diagnosing liver disease.

### Balancing the Output variable

The **SMOTE (Synthetic Minority Over-sampling Technique)** is a popular method used in machine learning to address the problem of imbalanced datasets. 

#### Importance Of Balancing

1. **Imbalanced Datasets**:
   - In many real-world datasets, the classes are imbalanced, meaning one class (typically the minority class) has significantly fewer instances than the other class (the majority class). This imbalance can cause machine learning models to become biased towards the majority class.

2. **Improving Model Performance**:
   - Models trained on imbalanced datasets may have poor performance in identifying the minority class, leading to high false negatives. SMOTE helps in creating a more balanced distribution of classes, thus improving the models ability to learn and predict both classes more effectively.

#### How SMOTE Works:

1. **Synthetic Sample Generation**:
   - SMOTE generates synthetic samples of the minority class by interpolating between existing minority class instances. Instead of simply duplicating the minority class instances, SMOTE creates new instances that are similar but not identical, which helps in providing a better decision boundary.

2. **Balancing the Dataset**:
   - By adding synthetic samples to the minority class, SMOTE balances the dataset, making it more suitable for training machine learning models.

By using SMOTE, we can effectively tackle the problem of class imbalance, leading to more robust and accurate models. 

```python
from collections import Counter# importing counter to check count of each label
from imblearn.over_sampling import SMOTE #for balancing the data
sm=SMOTE()#object creation
print(Counter(y))# checking count for each class 
X_sm,y_sm=sm.fit_resample(X,y)#applying sampling on target variable 
print(Counter(y_sm))# checking count after sampling for  each class
```

### Project Report: Challenges and Techniques Used in Liver Disease Prediction

#### 1. Introduction
This project aims to develop predictive models for diagnosing liver disease using various machine learning algorithms. Throughout the data preprocessing and model development stages, several challenges were encountered, and specific techniques were employed to address these issues effectively.

#### 2. Challenges Faced and Techniques Used

##### 2.1 Handling Outliers
**Challenge**: 
- The dataset contained numerous outliers, which could potentially distort the model's performance and lead to inaccurate predictions. Outliers can be caused by measurement errors, data entry errors, or genuine extreme values.

**Technique Used**:
- **IQR Technique**: The Interquartile Range (IQR) method was used to handle outliers. The IQR is the range between the 25th percentile (Q1) and the 75th percentile (Q3) of the data. Outliers are typically defined as data points that fall below Q1 - 1.5*IQR or above Q3 + 1.5*IQR.
- **Reason**: The IQR method is effective for identifying and handling outliers without assuming any distribution of the data. It is robust and less influenced by extreme values compared to other methods like z-scores.



**Outcome**:
- Despite using the IQR technique, it was not feasible to handle every outlier due to their high number. However, this method helped in reducing the impact of extreme values, resulting in more stable model performance.

##### 2.2 Model Overfitting in Random Forest
**Challenge**:
- The Random Forest model showed an accuracy of 100% on the training data, but only 82% on the testing data, indicating overfitting. Hyperparameter tuning did not improve the test accuracy, which remained at 82%.

**Technique Used**:
- **Hyperparameter Tuning**: Various hyperparameters such as the number of trees, maximum depth, and minimum samples split were tuned to find the optimal settings.
- **Reason**: Hyperparameter tuning helps in finding the best parameters that prevent overfitting and improve the model's generalization ability.


**Outcome**:
- Despite tuning, the test accuracy remained at 82%. Given its highest F1 and ROC AUC scores among all models, the Random Forest classifier was chosen as the final model. It provided the best balance between precision and recall, ensuring reliable predictions.
